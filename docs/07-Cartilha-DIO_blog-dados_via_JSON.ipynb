## ***2. Aplicando o conceito de JSON*** 

### ***2.1. Algumas adições***

```python
# implementando um teste:
class Foo(BaseModel): 
    bar: str
    message: str

# criando um endpoing teste
@app.get('/foobar/', response_model= Foo)
def foobar() -> Foo:
    return {'bar': "foo", "message": "Hello, Word"}
```

Por que isso é interessante? Em algum projeto, é possível que você tenha que fazer uma checagem de tipo. Fazendo algo assim:

```python
# implementando um teste:
class Foo(BaseModel): 
    bar: str
    message: str

# criando um endpoing teste
@app.get('/foobar/', response_model= Foo)
def foobar() -> dict[str, str]:
    return {'bar': "foo", "message": "Hello, Word"}
```

Se fizer isso, o resultado do endpoint será o mesmo que sem essa alteração. O interessante é lembrarmos que, por prescedencia, o 'response_model' é mais importante do que o retorno que a função dará '-> Foo' que utilizamos ou o '-> dict[str, str]' que alteramos.

Existem vários benefícios de utilizar o 'response_model', inclusive é aplicada uma documentação do FastAPI automaticamente ao utiliza-lo.
