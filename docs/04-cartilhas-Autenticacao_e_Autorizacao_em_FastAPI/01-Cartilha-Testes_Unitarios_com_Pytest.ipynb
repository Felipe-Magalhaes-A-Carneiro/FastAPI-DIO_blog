# ***FastAPI - Testes Unitários com Pytest***

----------------------------------------------------------------------------------------------------

Podemos criar testes unitários de manteiras síncrona e assíncrona. Nosso projeto é feito para a segunda forma, como sabemos.

Sendo assim, podemos realizar esses testes por meio do 'Pytest', uma biblioteca resposnsável por esse processo. Portanto utilizaremos as seguintes bibliotecas:

- pytest (utilizado para a execução de testes)
- pytest-asyncio (dará suporte aos testes junto com o pytest);
- httpx (necessário para utilizar a classe 'text client' que é retornado ao fazer uma conexão assíncrona);
- pytest-mock (colocar dados 'mocar' dados necessários)

E por recomendação, mas não obrigatório:

- asyncio_mode (facilita a escrita do teste)

# ***1. Reestruturando o projeto para realização de testes*** 

## ***1.1 Criação da pasta 'tests'***

Basta criar uma pasta de nome 'tests', na pasta principal do seu projeto.

## ***1.2 Corrigindo a estrutuda dos códigos dentro da pasta 'src'***

As pastas app, controllers, core, models, schemas, services, e views serão agora armazenadas na pasta 'scr' que deve ser criada na pasta principal do seu projeto.
