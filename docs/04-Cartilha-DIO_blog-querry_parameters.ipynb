# ***1. Criando duas rotas diferentes para acessos de dados distintos*** 

Criaremos duas rotas diferentes, uma acessa os posts de um database e o outro os posts que são feitos para o framework.

Vamos simular um acesso de um data base. No qual ele tenha as informações do /posts/{framework}.

Criaremos um novo método para um novo path, que se chamará apenas 'posts' e armazenaremos os dados que queremos que retorne na variavel de teste 'fake_db':

```python
# Criando uma simulação de data base
fake_db = [
    {'title': 'Criando uma aplicação com Django' , 'date': datetime.now(UTC)},
    {'title': 'Internacionalizando uma app FastAPI' , 'date': datetime.now(UTC)},
    {'title': 'Internacionalizando uma app Flask' , 'date': datetime.now(UTC)},
    {'title': 'Internacionalizando uma app Starlett' , 'date': datetime.now(UTC)}

]

#Criando novo método
@app.get('/posts')
def read_posts():
    return fake_db


# Criando rota
@app.get('/posts/{framework}')  
def read_framework_posts(framework: str): 
    return {
        "posts": [
                {'title': f'Criando uma aplicação com {framework}' , 'date': datetime.now(UTC)},
                
                {'title': f'Internacionalizando uma app {framework}' , 'date': datetime.now(UTC)}
            
            ]
    } # retorna uma lista de posts
```

Agora, o endereço http://127.0.0.1:8000/posts retorna o que esta no fake_db, enquanto o endereço http://127.0.0.1:8000/fastapi retorna informações que demos nas aulas anteriores, portanto, criamos dois tipos de porta: a 'posts' e a 'fastapi' ou qualquer outro framework que você queira colocar

# ***2. Criando os parâmetros para pesquisa*** 

https://fastapi.tiangolo.com/tutorial/query-params/

Vamos definir parâmetros 'default' para a pesquisa.

```python
#Criando novo método
@app.get('/posts')
def read_posts(skip: int = 0, limit: int = len(fake_db)): # mudamos os parâmetros do querry aqui
    return fake_db
```

- 'skip' é o desvio; 
- 'limit' é o limite de vezes que ocorre o querry

Segundo documentação, uma vez mudado os parâmetros, afetará as rotas seguintes. No nosso caso, trata-se da : http://127.0.0.1:8000/fastapi

Ao tentar acessa-la depois dessa mudança, verá a seguinte mensagem:

`{"detail":"Not Found"}`

Porém a porta http://127.0.0.1:8000/posts funciona normalmente.

```python
#Criando novo método
@app.get('/posts')
def read_posts(skip: int = 0, limit: int = len(fake_db)):
    return fake_db[skip: skip + limit]  # nesta parte
```

## ***3.1. Boolean - Utilizando o tipo de parâmetro*** 

É possível definirmos parâmetros default como 'boolean'. True ou False (Existem outros que o FastAPI também aceita, veja na documentação).

Vamos atualizar o nosso código:

```python
@app.get('/posts')
def read_posts(skip: int = 0, limit: int = len(fake_db), published: bool = True): # adicionamos o parâmetro bool 'published'
    return [post for post in fake_db[skip: skip + limit] if post["published"] is published] # filtramos o querry por meio do estado 'published', se ele for True, o FastAPI retornará, caso o contrário não.
```

Além disso, é necessário adiconarmos o estaro 'published' nos dados que estão no 'fake_db':

```python
# Criando uma simulação de data base
fake_db = [
    {'title': 'Criando uma aplicação com Django' , 'date': datetime.now(UTC), "published": True},
    {'title': 'Internacionalizando uma app FastAPI' , 'date': datetime.now(UTC), "published": True},
    {'title': 'Internacionalizando uma app Flask' , 'date': datetime.now(UTC), "published": True},
    {'title': 'Internacionalizando uma app Starlett' , 'date': datetime.now(UTC), "published": True}
]
```

# ***4. Boas práticas - Os argumentos obrigatórios, devem ser declarados inicialmente, como sendo os primeiros*** 

Logo, no nosso caso, ficará assim:

```python
#Criando novo método
@app.get('/posts')
def read_posts(published: bool, skip: int = 0, limit: int = len(fake_db)): # 'published' passa ser o primeiro parâmetro já que ele é obrigatório
    return [post for post in fake_db[skip: skip + limit] if post["published"] is published]
```

Nota: Perceba que por ele ser obrigatório, não há necessidade de definir como `published: bool = True`, basta utilizarmos o `published: bool`. Desta forma, o FastAPI entende que ele é obrigatório.
