    poetry add fastapi

## ***1.4. Selecionar Interpretador***

Copiaremos o caminho dado pelo poetry env info:

    Path:           C:\Users\Felipe\AppData\Local\pypoetry\Cache\virtualenvs\fastapi-dio-blog-w_zijIWS-py3.14

Em seguida, no VScode, utilize o atalho:

 Ctrl + Shift + P

Na parte superior da tela, aparecerá um campo, escreveremos 

    Python: Select Interpreter

Em seguida escolhemos a opcão: "Enter interpreter path..."

Colamos o resultado do Path info que adquirimos anteriormente neste campo: 

    Path:           C:\Users\Felipe\AppData\Local\pypoetry\Cache\virtualenvs\fastapi-dio-blog-w_zijIWS-py3.14

Perceba que um arquivo '.vscode' será criado automáticamente no seu projeto.
