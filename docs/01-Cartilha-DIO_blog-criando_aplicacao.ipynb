# ***1. Instalação e Configuração do FastAPI:*** 

https://fastapi.tiangolo.com/#installation

Em 'main.py', traremos:

```python
from fastapi import FastAPI

# instanciando a classe FastAPI
app = FastAPI()
```

```python
# Criando rota
@app.get('/')  # definimos o verbo HTTP que vamos usar
def read_root():
    return {"message": "Hello, Word!"}
```

Rodando a aplicação

No terminal, digite:

    uvicorn main:app --reload

- main = nome do arquivo;
- app = nome da variável;
- --reload = toda vez que houver uma mudança no código, ele irá recarregar automáticamente

Obs.: Caso você tenha colocado o arquivo 'main.py' dentro de uma pasta 'app' ou qualquer outro tipo, como é o meu caso, faça assim:

    uvicorn *nome_do_arquivo*.main:app --reload

No meu caso ficou:

    uvicorn app.main:app --reload

- app = nome da pasta que esta o arquivo 'main';
- main = nome do arquivo;
- app = nome da variável;
- --reload = toda vez que houver uma mudança no código, ele irá recarregar automáticamente

# ***2. Como reinicializar a API?*** 

Caso a API feche ou você tenha desligado o PC, faça o seguinte:

Caso você precise acessar a sua API novamente, utilize outra vez no terminal o comando:

`main:app --reload`

***Obs.:*** Lembre-se que, no meu caso, foi necessário usar o comando `app.main:app --reload` com a pequana adição do 'app' no começo do comando, pois o meu arquivo 'main' esta localizado na pasta 'app', então ele deve primeiro acessar a pasta 'app' e então o arquivo 'main.py', caso o contrário dará ERRO.
