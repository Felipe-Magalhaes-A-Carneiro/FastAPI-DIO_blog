# ***1. Path Parameters - FastAPI:*** 

https://fastapi.tiangolo.com/tutorial/path-params/

São parâmetros que vem nos caminhos/endereço da requisição/URL

No momento, no nosso arquivo main.py é a seguinte parte do código:

```python
# Criando rota
@app.get('/')  # definimos o verbo HTTP que vamos usar
```

Vamos mudar:

```python
# Criando rota
@app.get('/posts/{framework}')  # definimos o verbo HTTP que vamos usar
```

O termo 'framework' usado não é obrigatório, poderia ser qualquer nome. É recomendo um no qual ajuda a entender do que se trata o posts, uma mesma lógica do nome que devemos dar as variáveis.

# ***2. Criando uma série de posts com datetime*** 

Os posts que faremos, retornará uma lista com um datetime que nos ajudará a entender melhor o processo de criação, por meio do dia e horário que ocorrerem.

```python
@app.get('/posts/{framework}')  # definimos o verbo HTTP que vamos usar
def read_posts(framework):
    return {
        "posts": [
                {'title': f'Criando uma aplicação com {framework}' , 'date': datetime.now(UTC)},
                
                {'title': f'Internacionalizando uma app {framework}' , 'date': datetime.now(UTC)}
            
            ]
    } # retorna uma lista de posts
```
