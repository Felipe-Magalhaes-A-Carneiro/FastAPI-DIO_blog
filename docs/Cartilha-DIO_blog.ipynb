# 1. Instalação: 

https://fastapi.tiangolo.com/#installation

Vá no Terminal e digite:

    `pip install "fastapi[standard]"`

# 2. Poetry init: 

Vá no Terminal e digite:

    `poetry init`

Uma série de informações irão aparecer para você defini-las, no meu caso fiz assim:

> Package name [formacao_python-fundamentals]:  
Version [0.1.0]:  
Description []:  Blog - Meu primeiro projeto com FastAPI

License []:  
Compatible Python versions [>=3.14]:
