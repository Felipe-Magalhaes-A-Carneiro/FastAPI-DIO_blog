# ***DIO Blog - Primeiro projeto com FastAPI***

----------------------------------------------------------------------------------------------------

# ***1. Instalação:*** 

https://fastapi.tiangolo.com/#installation

Vá no Terminal e digite:

    `pip install "fastapi[standard]"`

----------------------------------------------------------------------------------------------------

# ***2. Poetry init:***

Vá no Terminal e digite:

    `poetry init`

Uma série de informações irão aparecer para você defini-las, no meu caso fiz assim:

# ***5. Selecionar Interpretador***

Copiaremos o caminho dado o poetry env info:

    Path:           C:\Users\Felipe\AppData\Local\pypoetry\Cache\virtualenvs\fastapi-dio-blog-w_zijIWS-py3.14

Em seguida, Ctrl + Shift + P

Escreveremos "Python: Select Interpreter"

Em seguida escolhemos a opcão: "Enter interpreter path..."

Colamos o resultado do Path nele: 

    Path:           C:\Users\Felipe\AppData\Local\pypoetry\Cache\virtualenvs\fastapi-dio-blog-w_zijIWS-py3.14

Perceba que um arquivo '.vscode' será criado automáticamente.
